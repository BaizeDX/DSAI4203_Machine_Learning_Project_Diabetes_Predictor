# Notebook 13.5: Overall Experiment Summary

This notebook provides the final high-level summary of the project after the full methodology evolution. Unlike Notebook 11.5, which serves as an interim phase summary, this notebook looks across the complete experiment path.


## Experiment Progression

The project developed through four clear stages:
- **Exploration and baselines**: EDA, simple baselines, and early model-family exploration clarified the shape of the problem.
- **Boosting-focused search**: default XGBoost and LightGBM baselines, followed by a sequence of XGBoost refinements, identified boosting as the strongest practical direction.
- **Validation redesign**: cross-validation and later validation-focused supporting analyses showed that evaluation reliability mattered as much as raw model tuning.
- **Final reporting and synthesis**: the project narrative was reorganized around the most defensible methodological choices rather than around isolated score improvements.


## Major Turning Points

| Turning point | Why it mattered |
| --- | --- |
| Moving from simple baselines to boosting | Established that the task benefits from stronger nonlinear tabular models |
| Comparing default XGBoost and default LightGBM | Confirmed that the model-family decision should be made under aligned preprocessing and split logic |
| Strengthening validation in Notebook 11 | Replaced a fragile holdout-centered workflow with a more trustworthy evaluation framework |
| Re-examining validation design in Notebooks 12 and 13 | Showed that leaderboard-facing and cross-validation-facing decisions both affect how results should be interpreted |

These turning points made the final approach stronger not only in score, but also in methodological credibility.


## Final Method Preference

A practical interpretation of the full project is that the preferred final direction combines:
- a boosting-based tabular model family
- careful retention of useful original and engineered features
- a validation protocol chosen for reliability rather than convenience
- report-ready interpretation that distinguishes exploratory gains from more trustworthy gains

The strongest contribution of the later notebooks is therefore not a single isolated trick, but the combination of model quality and evaluation discipline.


## Final Takeaways

Based on the current experiments, the project supports three broad conclusions:
- Tree-based boosting remained the most dependable modeling direction for this dataset.
- Validation design materially affected which scores were trustworthy enough to guide decisions.
- The final notebook sequence is best understood as a methodological progression from exploration to defensible evaluation rather than as a simple score-chasing process.
